```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

print("Starting Orders ETL Pipeline")

data = [
    ("C001", "Laptop", 1200),
    ("C002", "Phone", 800),
    ("C003", "Tablet", 400),
    ("C001", "Mouse", 50)
]

df = spark.createDataFrame(data, ["customer_id", "product", "price"])

print("Raw Data")
df.show()

# Data validation
if df.count() == 0:
    raise Exception("Dataset is empty")

# Transformation
orders_summary = df.groupBy("customer_id").sum("price")
orders_summary = orders_summary.withColumnRenamed("sum(price)", "total_price")

print("Orders Summary")
orders_summary.show()

# Save table
orders_summary.write.mode("overwrite").saveAsTable("orders_summary_table")

print("Orders ETL Pipeline Completed")
```
